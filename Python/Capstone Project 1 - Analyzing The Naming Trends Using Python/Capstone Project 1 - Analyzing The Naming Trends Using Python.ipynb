```python
arr=[]

for year in range(1880,2022):
    source = r'C:\Users\dutta\Dropbox\Documents\MNIT\Python\Assignments\Capstone Project 1 - Analyzing The Naming Trends Using Python\names\yob%d.txt' %year
    frame = pd.read_csv(source, names = ['Name','Gender','Count'])
    frame['Year'] = year
    arr.append(frame)

names_data = pd.concat(arr, ignore_index=True)
names_data.shape
```

```python
names_data.head(50)
```

```python
names_data.tail(50)
```

```python
names_data.groupby(['Year','Gender']).Count.sum()
```

```python
births = names_data.pivot_table('Count', index = 'Year', columns = 'Gender', aggfunc = sum)
births
```

```python
births.plot(title='Total Birth by Gender and Year', ylabel = 'Count')
```

```python
def pname(popular_name):
    popular_name['Popular Names'] = ((popular_name['Count'] / popular_name['Count'].sum()) * 100)
    return popular_name

names_data_popular = names_data.groupby(['Year', 'Gender']).apply(pname)
names_data_popular
```

```python
def top_name(popular_name):
    return popular_name.sort_values(by='Count', ascending = False)[:10]

p_name = names_data_popular.groupby(['Gender','Count'])
top_100 = p_name.apply(top_name)
top_100.reset_index(inplace=True, drop=True)

M_name = top_100[top_100['Gender'] == 'M']
F_name = top_100[top_100['Gender'] == 'F']
```
