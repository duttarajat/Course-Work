# Module 9 - Unsupervised Learning - Case Study 2 - Unsupervised Learning

```python
import numpy as np
import pandas as pd
from copy import deepcopy
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
```

```python
cars = pd.read_csv('cars-1.csv')
cars.head()
```

```python
car_features = cars[['mpg', 'disp', 'hp']]
car_features.head()
```

```python
car_features_stan = pd.DataFrame(data = (StandardScaler().fit_transform(car_features)), columns = ['mpg', 'disp', 'hp'])
car_features_stan.head()
```

```python
kmns3 = KMeans(n_clusters=3)
kmns3.fit(car_features_stan)
```

```python
car_features['Cluster3'] = kmns3.fit_predict(car_features_stan)
car_features.head()
```

```python
cars['Cluster3'] = kmns3.fit_predict(car_features_stan)
cars.head()
```

```python
cars[cars['Cluster3'] == 0]
```

```python
cars[cars['Cluster3'] == 1]
```

```python
cars[cars['Cluster3'] == 2]
```

```python
kmns5 = KMeans(n_clusters=5)
kmns5.fit(car_features_stan)
```

```python
car_features['Cluster5'] = kmns5.fit_predict(car_features_stan)
car_features.head()
```

```python
cars['Cluster5'] = kmns5.fit_predict(car_features_stan)
cars.head()
```
