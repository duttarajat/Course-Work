# Module 9 - Unsupervised Learning - Assignment 1 - K Means (Suboptimal)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('customers.csv')
data.head()
```

```python
data.isna().sum()
```

```python
#There are no cells with null values
```

```python
X = data['Age']
Y = data['Spending Score (1-100)']
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data = data, x = X, y = Y)
plt.title('Age Vs Spending Score')
plt.show()
```

```python
kmns = KMeans(n_clusters= 3)
kmns.fit(data[['Age', 'Spending Score (1-100)']])
centroids = kmns.cluster_centers_
data['Clusters'] = kmns.labels_

plt.figure(figsize=(8,5))
sns.scatterplot(data = data, x = X,  y = Y, hue = 'Clusters', palette = 'rainbow')
sns.scatterplot(x = centroids[:,0], y = centroids[:,1], marker = '*', size = 1, sizes= [500], palette = 'rainbow', hue = range(3), legend = None)
plt.show()
```
