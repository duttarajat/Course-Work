```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
```

```python
wine = pd.DataFrame(data = load_wine().data, columns = load_wine().feature_names)
wine.head()
```

```python
wine.shape
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA, FactorAnalysis as FA
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
pca = PCA(n_components = 2)
wine_pca = pd.DataFrame(data = (pca.fit_transform(wine)), columns = ['PCA1', 'PCA2'])
```

```python
pca.explained_variance_ratio_
```

```python
fa= FA(n_components = 2)
wine_fa = pd.DataFrame(data = (fa.fit_transform(wine)), columns = ['FA1', 'FA2'])
```

```python
X = wine
Y = load_wine().target
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.75, random_state = 0)
lda = LDA(n_components = 2)
X_train = pd.DataFrame(data = (lda.fit_transform(X_train, Y_train)), columns = ['LDA1', 'LDA2'])
X_test = pd.DataFrame(data = (lda.fit_transform(X_test, Y_test)), columns = ['LDA1', 'LDA2'])
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
wine_lda = pd.concat([X_train, X_test], ignore_index = True)
```
