```python
import pandas as pd
import matplotlib.pyplot as plt

churn = pd.read_csv('customer_churn.csv')
```

```python
C_type = list(churn['Contract'].value_counts().sort_index().index)
Count = list(churn['Contract'].value_counts().sort_index())
plt.bar(C_type, Count, color = 'orange')
plt.title('Distribution of Contract')
plt.xlabel('Contract Type of Customer')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(churn['MonthlyCharges'], color = 'forestgreen')
plt.title('Distribution of Monthly Charges')
plt.xlabel('Monthly Charges Incurred')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.scatter(churn['tenure'], churn['TotalCharges'], color = 'indigo')
plt.xlabel('Tenure of the customer')
plt.ylabel('Total Charges Incurred')
plt.title('Total Charges vs Tenure')
plt.show()
```

```python
c_pm_ec = churn[churn['PaymentMethod'] == 'Electronic check']
c_pm_mc = churn[churn['PaymentMethod'] == 'Mailed check']
c_pm_bt = churn[churn['PaymentMethod'] == 'Bank transfer (automatic)']
c_pm_cc = churn[churn['PaymentMethod'] == 'Credit card (automatic)']
c='olive'
plt.boxplot([c_pm_ec['MonthlyCharges'],c_pm_mc['MonthlyCharges'],c_pm_bt['MonthlyCharges'],c_pm_cc['MonthlyCharges']],notch=True,patch_artist=True,boxprops=dict(facecolor=c,color=c),capprops=dict(color=c),whiskerprops=dict(color=c),flierprops=dict(color=c,markeredgecolor=c),medianprops=dict(color=c))
plt.xlabel('Payment Method of Customer')
plt.ylabel('Monthly Charges Incurred')
plt.title('Monthly Charges vs. Payment Method')
plt.xticks([1,2,3,4],['Electronic Check', 'Mailed Check', 'Bank Transfer', 'Credit Card'])
plt.show()
```
