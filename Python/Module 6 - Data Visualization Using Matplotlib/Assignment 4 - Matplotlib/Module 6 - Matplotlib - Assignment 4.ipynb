```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cars = pd.read_csv('cars-4.csv')
```

```python
Model = list(cars['model'].value_counts().sort_index().index)
Cyl = list(cars['cyl'].values)
plt.figure(figsize=(10,10))
plt.pie(Cyl, labels = Model)
plt.show()
```
