```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cars = pd.read_csv('cars-5.csv')
```

```python
plt.stackplot(list(cars['am'].values), list(cars['carb'].values))
plt.xlabel('Transmission')
plt.ylabel('Number of carburetors')
plt.title('Transmission vs Number of carburetors')
plt.xticks([0,1],['Manual','Automatic'])
plt.show()
```
